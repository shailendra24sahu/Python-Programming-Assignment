Question 1:
Please write a program using generator to print the numbers which can be divisible by 5 and
7 between 0 and n in comma separated form while n is input by console.
Example:
- If the following n is given as input to the program:
100
Then, the output of the program should be:
0,35,70

```python
def divisible_by_5_and_7(n):
    for i in range(n+1):
        if i % 5 == 0 and i % 7 == 0:
            yield i

n = int(input("Enter a number: "))
result = divisible_by_5_and_7(n)

print(*result, sep=', ')
```

the program prints the numbers generated by the divisible_by_5_and_7 generator function in comma-separated form using the print() function with the * operator to unpack the generator object and the sep parameter set to ', '.

Question 2:
Please write a program using generator to print the even numbers between 0 and n in comma
separated form while n is input by console.

```python
def even_numbers(n):
    for i in range(n+1):
        if i % 2 == 0:
            yield i

n = int(input("Enter a number: "))
result = even_numbers(n)

print(*result, sep=', ')
```

Q3. Please write a program using list comprehension to print the Fibonacci Sequence in comma
separated form with a given n input by console.
Example:
If the following n is given as input to the program:
7

Then, the output of the program should be:
0,1,1,2,3,5,8,13

```python
n = int(input("Enter a number: "))

fib = [0, 1]
[fib.append(fib[-2] + fib[-1]) for i in range(n+1)]

print(*fib[:n+1], sep=', ')
```

Question 5:
Define a class named Shape and its subclass Square. The Square class has an init function
which takes a length as argument. Both classes have a area function which can print the area
of the shape where Shape&#39;s area is 0 by default.

```python
class Shape:
    def __init__(self):
        self.area = 0

    def get_area(self):
        return self.area

class Square(Shape):
    def __init__(self, length):
        super().__init__()
        self.length = length
        self.area = self.length ** 2
```

```python
shape = Shape()
print(shape.get_area()) 

square = Square(5)
print(square.get_area()) 
```
